```python
import pandas as pd
import numpy as np
from random import shuffle
from copy import deepcopy
from pathlib import Path
```

```python
data_dir = '../Data-Populated-Unfiltered/Folds/'
save_dir = '../Data-Populated-Unfiltered/Sites/'
datasets = ['C2C12', 'ESC', 'MLC', 'NBC', 'L1-I','L1-F','L1-R','L6']
clss = ['MAPK1', 'MTOR']
num_folds = 5
k_folds = 5

Path(save_dir).mkdir(parents=True, exist_ok=True)
```

```python
data_fp = ['{}{}_{}_FLD.csv'.format(data_dir, dat,cl) for dat in datasets for cl in clss ]
class_columns = ['y.{}'.format(c) for c in clss]
fold_columns = ['F{}'.format(val+1) for val in range(num_folds)]
motifs = ['{}_motif'.format(dat) for dat in datasets]
```

```python
data_fp = ['{}{}_{}_FLD.csv'.format(data_dir, dat,cl) for dat in datasets for cl in clss ]
save_fp = ['{}{}_{}_FLD.csv'.format(save_dir, dat,cl) for dat in datasets for cl in clss ]
```

```python
data = []

for fp in data_fp:
    data.append(pd.read_csv(fp))
```

```python
for df, save_f in zip(data, save_fp):
    df[['site'] + fold_columns + class_columns].to_csv(save_f, index=False)
```
